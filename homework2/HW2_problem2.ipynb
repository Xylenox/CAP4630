```
import numpy as np
from mpl_toolkits import mplot3d
%matplotlib inline
import matplotlib.pyplot as plt
```

```
m = 100

np.random.seed(42)

x1 = np.random.rand(m, 1)
x2 = np.random.rand(m, 1)

xs = np.column_stack((x1, x2))
ys = x1 + x2
print(xs.shape)
print(ys.shape)
```

```
ax = plt.axes(projection='3d')
ax.scatter3D(x1, x2, ys)
```

# Linear regression using numpy


```
# number of epochs
epochs = 20
# learning rate
lr = 0.01
# batch size
batch_size = 4
```

```
# initial value for weight w and bias b
w1 = np.random.randn(1)
w2 = np.random.randn(1)
w = np.random.rand(3, 1)
b = np.zeros(1)
xb = np.column_stack((np.ones(m), xs))

print(w.shape)
print(ys.shape)
print(xb.dot(w).shape)
print(xb.T.dot(xb.dot(w)).shape)
```

```
for epoch in np.arange(epochs):
  shuffle = np.random.permutation(m)
  shuffle_x = xb[shuffle]
  shuffle_y = ys[shuffle]
  for i in range(0, m, batch_size):
    sample_x = shuffle_x[i:i+batch_size]
    sample_y = shuffle_y[i:i+batch_size]
    w = w - 1 / m * lr * sample_x.T.dot(sample_x.dot(w) - sample_y)
  print(w)
```

```
m_test = 100

np.random.seed(42)

x_test1 = np.random.rand(m_test, 1)
x_test2 = np.random.rand(m_test, 1)

x_tests = np.column_stack((x_test1, x_test2))
y_tests = 3 * x1 + 4 * x2 + 5 + 0.25 * np.random.rand(m, 1)

x_testb = np.column_stack((np.ones(m), x_tests))
```

```
test_loss = 0
for i in np.arange(m):
  test_loss += 0.5 * ((x_testb[i].dot(w)).sum() - y_tests[i]) ** 2
print((x_testb[0]))
print(w.squeeze())
print((x_testb[0]).dot(w))
test_loss /= m_test
```
